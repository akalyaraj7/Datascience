```python
import tensorflow as tf
import numpy as np
```

```python
# Four arm bandit and it's distribution of the bandits 
bandits = [0,2,0,-0.2,-5]
num_bandits = len(bandits)
```

```python
def pullBandit(bandit):
    """
    Returns the rewards of a Bandit algorithm
    """
    result = np.random.randn(1)
    
    if result > bandit:
        return 1
    else:
        return -1
```

```python
## The Agent
```

```python
tf.reset_default_graph()
```

```python
#initialize all weights to 1
weights = tf.Variable(tf.ones([num_bandits]))
#Get the maximum rewarded bandits
choosen_action = tf.argmax(weights, 0)
```

```python
# Agent
## Poicy loss equation: Loss = -log(π)*A , π is the Policy, A - how much the action is better than the baseline
```

```python
# Declare variables to hold rewards and actions
reward_holder = tf.placeholder(shape=[1], dtype=tf.float32)
action_holder = tf.placeholder(shape=[1], dtype=tf.int32)
#calculate the plolicy value (weights here)
responsible_weight = tf.slice(weights, action_holder, [1])
# calculate the loss
loss = -(tf.log(responsible_weight) * reward_holder)
# instantiate the optimizer  and minimize the loss
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.001)
update = optimizer.minimize(loss)
```

```python
## Training the Agent 
```

```python
total_episodes = 1000

total_reward = np.zeros(num_bandits)

#using epsilon greedy algorithm here
e = 0.1
```

```python
init = tf.initialize_all_variables()
```

```python
with tf.Session() as sess:
    sess.run(init)
    i = 0
    while i < total_episodes:
        #Choose either a random action or one from our network
        if np.random.rand(1) < e: # exploration
            action = np.random.randint(num_bandits)
        else: # exploitation
            action = sess.run(choosen_action)
        
        # Get reward from picking one of the bandits
        reward = pullBandit(bandits[action])
        
        # Update the network
        _, resp, ww = sess.run([update,responsible_weight,weights], feed_dict={reward_holder:[reward],action_holder:[action]})
        
        # update running tally of scores 
        total_reward[action] += reward
        if i%50:
            print(f"Running reward for the {str(num_bandits)} bandits : {str(total_reward)}")
        i+=1
    
print(f"Agent thinks bandit {str(np.argmax(ww)+1)} is the most promising")

# Checking if the agent was right or not

if np.argmax(ww) == np.argmax(-np.array(bandits)):
    print ("...and it was right!")
else:
    print ("...and it was wrong!")
```
